```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
url = 'https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/1_linear_reg/homeprices.csv'
df = pd.read_csv(url)
df.head()
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color='blue',marker="*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
url = 'https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/2_linear_reg_multivariate/homeprices.csv'
df = pd.read_csv(url)
df
```

```python
#Cal Median
#df.bedrooms.median()
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

price = m1 * area + m2 * bedrooms + m3 * age + b
indep - price
m1 and m2 and m3 - coeff
area and ebd and age - indep
b = intercept

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#reg.predict([[3500,3,40]])
model.predict(5000)
```
